```python

def categorize_title(title):
    title = title.upper()

    # Check for gaming laptops
    if 'GAMING LAPTOP' in title:
        return 'Gaming Laptop'
    # Check for regular laptops
    elif 'LAPTOP' in title:
        return 'Laptop'
    # Check for headphones or tai nghe
    elif 'TAI NGHE' in title or 'HEADPHONE' in title:
        return 'Tai nghe'
    # Default for other products
    else:
        return 'Khác'

# Assuming you have assigned the list to the 'title' column
df['title'] = df['name']  # or if the original column name is 'name'

# Apply the classification function
df['category'] = df['title'].apply(categorize_title)

# Print the first 5 results
print(df.head(5))

```

```python
from sentence_transformers import SentenceTransformer
from typing import List

embedding_model = SentenceTransformer("keepitreal/vietnamese-sbert")

def get_embedding(text: str) -> List[float]:
    if not text.strip():
        print("Attempted to get embedding for empty text.")
        return []

    embedding = embedding_model.encode(text)
    return embedding.tolist()
```

```python
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
from chromadb.utils import embedding_functions
import re
import json
import os

# Khởi tạo embedding function với model SBERT tiếng Việt
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="keepitreal/vietnamese-sbert"
)

os.makedirs("databaseQQ/VECTOR_STORE", exist_ok=True)

# Khởi tạo ChromaDB client
client = chromadb.PersistentClient(path='databaseQQ/VECTOR_STORE')


# Recreate the collections with the correct embedding function
collection1 = client.get_or_create_collection(
    name="pbl7_vgear1",
    embedding_function=sentence_transformer_ef
)

collection2 = client.get_or_create_collection(
    name="pbl7_vgear2",
    embedding_function=sentence_transformer_ef
)

# Tiền xử lý văn bản
def preprocess_text(text: str) -> str:
    text = text.lower()
    text = re.sub(r'[^\w\s,.]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Hàm chung để thêm sản phẩm vào collection
def add_product_to_collection(product, product_id, collection, fields_to_combine):
    combined_text = " ".join([str(product[field]) for field in fields_to_combine if field in product])
    combined_text = preprocess_text(combined_text)

    if not combined_text:
        print(f"Empty processed text for product ID {product_id}")
        return

    # Kiểm tra và chuẩn hóa trường images thành danh sách các chuỗi (nếu chưa là list)
    raw_images = product.get("images", [])  # Correct field name to 'images'
    if isinstance(raw_images, str):
        try:
            images = json.loads(raw_images)  # Parse string to list if needed
        except json.JSONDecodeError:
            images = []  # Nếu không thể parse, gán ảnh rỗng
    elif isinstance(raw_images, list):
        images = raw_images  # Nếu đã là list, giữ nguyên
    else:
        images = []  # Nếu không phải string hoặc list, trả về ảnh rỗng

    # Thêm sản phẩm vào collection
    collection.add(
        ids=[str(product_id)],
        documents=[combined_text],
        metadatas=[{
            "title": product.get("title", ""),
            "description": product.get("description", ""),
            "selling_price": product.get("selling_price", ""),
            "images": json.dumps(images),  # Correctly store images as JSON string under 'images'
            "category": product.get("category", "")
        }]
    )


# Thêm sản phẩm từ DataFrame vào hai collection khác nhau
for index, product in df_chewy.iterrows():
    add_product_to_collection(product, index, collection1, fields_to_combine=["title", "description"])
    add_product_to_collection(product, index, collection2, fields_to_combine=["title", "category"])

# Truy vấn
results = collection1.query(
    query_embeddings=sentence_transformer_ef("tai nghe"),
    n_results=5
)

# Hiển thị kết quả
for i in range(len(results['documents'][0])):
    metadata = results['metadatas'][0][i]
    print(f"Metadata: {metadata}")  # This will show all metadata fields
    images = json.loads(metadata['images']) if metadata.get('images') else []

    print(f"Title: {metadata['title']}")
    print(f"Description: {metadata['description']}")
    print(f"Selling Price: {metadata['selling_price']}")
    print(f"Category: {metadata['category']}")
    print(f"Images: {images}")
    print()
```

```python
def weighted_reciprocal_rank(doc_lists):
        """
        Perform weighted Reciprocal Rank Fusion on multiple rank lists.
        You can find more details about RRF here:
        https://plg.uwaterloo.ca/~gvcormac/cormacksigir09-rrf.pdf

        Args:
            doc_lists: A list of rank lists, where each rank list contains unique items.

        Returns:
            list: The final aggregated list of items sorted by their weighted RRF
                  scores in descending order.
        """
        c = 60  # Parameter from the paper
        weights = [1] * len(doc_lists)  # Weights for each list

        if len(doc_lists) != len(weights):
            raise ValueError("Number of rank lists must equal the number of weights.")

        # Collect all unique documents
        all_documents = {doc["description"] for doc_list in doc_lists for doc in doc_list}
        rrf_score_dic = {doc: 0.0 for doc in all_documents}

        # Calculate RRF scores
        for doc_list, weight in zip(doc_lists, weights):
            for rank, doc in enumerate(doc_list, start=1):
                rrf_score = weight * (1 / (rank + c))
                rrf_score_dic[doc["description"]] += rrf_score

        # Sort documents by RRF scores in descending order
        sorted_documents = sorted(
            rrf_score_dic.keys(), key=lambda x: rrf_score_dic[x], reverse=True
        )

        # Map sorted content back to original documents
        page_content_to_doc_map = {
            doc["description"]: doc for doc_list in doc_lists for doc in doc_list
        }
        return [page_content_to_doc_map[content] for content in sorted_documents]


def hybrid_search(query: str, query_embedding: list, collection, limit=10):
    if query_embedding is None:
        return "Invalid query or embedding generation failed."

    # Perform vector search in ChromaDB
    vector_results = collection.query(
        query_embeddings=[query_embedding],
        n_results=limit
    )
    # Parse vector search results
    vector_results = [
        {
            "_id": vector_results['ids'][0][i],
            "title": vector_results['metadatas'][0][i].get('title'),
            "description": vector_results['metadatas'][0][i].get('description'),
            "selling_price": vector_results['metadatas'][0][i].get('selling_price'),
            "images": vector_results['metadatas'][0][i].get('images'),
            "category": vector_results['metadatas'][0][i].get('category'),
            "distance": vector_results['distances'][0][i]
        }
        for i in range(len(vector_results['ids'][0]))
    ]

    # Perform keyword search in ChromaDB
    keyword_results = collection.query(
        query_texts=[query],
        n_results=limit
    )

    keyword_results = [
        {
            "_id": keyword_results['ids'][0][i],
            "title": keyword_results['metadatas'][0][i].get('title'),
            "description": keyword_results['metadatas'][0][i].get('description'),
            "selling_price": keyword_results['metadatas'][0][i].get('selling_price'),
            "images": keyword_results['metadatas'][0][i].get('images'),
            "category": keyword_results['metadatas'][0][i].get('category'),
            "distance": keyword_results['distances'][0][i]
        }
        for i in range(len(keyword_results['ids'][0]))
    ]

    # Merge results and apply rank fusion
    doc_lists = [vector_results, keyword_results]
    fused_documents = weighted_reciprocal_rank(doc_lists)
    return fused_documents


def enhance_prompt(query: str, query_embedding: list, collection):
        get_knowledge = hybrid_search(query, query_embedding, collection, 10)
        print('hybrid_search_result:', get_knowledge)
        enhanced_prompt = ""
        print(len(get_knowledge))
        for result in get_knowledge:
            enhanced_prompt += f"Title: {result.get('title', 'N/A')}, Content: {result.get('description', 'N/A')}, Selling Price: {result.get('selling_price', 'N/A')}, Images: {result.get('images', 'N/A')}\n"
        return enhanced_prompt
```

```python
def display_result(result_str: str):
    # Parse the result string into structured data
    products = []
    for product_info in result_str.strip().split('\n'):
        product = {}
        for field in product_info.split(', '):
            if ': ' in field:
                key, value = field.split(': ', 1)
                product[key.strip().lower()] = value.strip()
        products.append(product)

    # Print the parsed product details
    for product in products:
        print(product.keys())  # Xem các key thực sự có trong dict
        print("Sản phẩm:", product.get("title"))
        print("Giá:", product.get("selling price"))
        print("Mô tả:", product.get("description"))
        print("Hình ảnh:", product.get("images"))
        print("----------")
```

```python
def search_products(query: str, collection, top_k: int = 5):
    # Từ điển ánh xạ category với các từ khóa
    keyword_filters = {
        'Laptop ASUS ExpertBook': ['Laptop ASUS ExpertBook', 'ASUS ExpertBook', 'B1 B1402CVA', 'NK1280'],
        'Laptop Lenovo IdeaPad': ['Laptop Lenovo IdeaPad', 'Lenovo IdeaPad', 'Slim 3 15ABR8', '82XM00MCVN'],
        'Laptop ASUS Vivobook': ['Laptop ASUS Vivobook', 'Vivobook', 'A1405ZA', 'KM264W'],
        'Laptop ASUS Zenbook': ['Laptop ASUS Zenbook', 'Zenbook', 'UM3406KA', 'PP555WS'],
        'Laptop ASUS Zenbook UX': ['Laptop ASUS Zenbook UX', 'Zenbook UX', '3407QA', 'QD299WS'],
        'Laptop ASUS Vivobook S': ['Laptop ASUS Vivobook S', 'Vivobook S', 'S5406SA', 'PP060WS'],
        'Laptop ASUS Zenbook OLED': ['Laptop ASUS Zenbook OLED', 'Zenbook OLED', 'UX3407QA', 'QD299WS'],
        'Laptop ASUS ZenBook UX425EA': ['Laptop ASUS ZenBook UX425EA', 'ZenBook UX425EA', 'KI439T'],
        'Laptop ASUS Zenbook Duo': ['Laptop ASUS Zenbook Duo', 'Zenbook Duo', 'UX482EG', 'KA166T'],
        'Laptop gaming Acer Predator': ['Laptop gaming Acer Predator', 'Acer Predator', 'Helios Neo 16', 'PHN16 72 78L4'],
        'Laptop gaming MSI Thin': ['Laptop gaming MSI Thin', 'MSI Thin', 'B13UC 2044VN'],
        'Tai nghe Asus ROG Pelta': ['Tai nghe Asus ROG Pelta', 'Asus ROG Pelta', 'WL RGB Black'],
        'Tai nghe Logitech G733': ['Tai nghe Logitech G733', 'Logitech G733', 'LIGHTSPEED Wireless', 'Black'],
        'Tai nghe Logitech G435': ['Tai nghe Logitech G435', 'Logitech G435', 'Lightspeed Wireless', 'Black']
    }

    # Tìm các từ khóa trong truy vấn của người dùng và lưu các category trùng khớp
    filters = []
    for category, keywords in keyword_filters.items():
        if any(keyword.lower() in query.lower() for keyword in keywords):
            filters.append(category)

    print("Filters found:", filters)  # Debug print

    # Chuẩn hóa truy vấn và tạo embedding
    query = query.lower().strip()
    query_embedding = get_embedding(query)

    # Thực hiện tìm kiếm vector
    vector_results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k,
    )

    # Lọc kết quả dựa trên từ khóa nếu có
    filtered_results = []
    for idx, metadata in enumerate(vector_results["metadatas"][0]):
        category = metadata.get("category", "").lower().strip()  # Lấy category của sản phẩm
        if filters:
            # Nếu có filter, kiểm tra xem category của sản phẩm có trong filters không
            if category in [f.lower().strip() for f in filters]:
                filtered_results.append({
                    "document": vector_results["documents"][0][idx],
                    "metadata": metadata
                })
        else:
            # Nếu không có filter, trả về toàn bộ kết quả từ vector search
            filtered_results.append({
                "document": vector_results["documents"][0][idx],
                "metadata": metadata
            })

    # Hiển thị kết quả tìm kiếm
    if filtered_results:
        for result in filtered_results:
            metadata = result["metadata"]
            print("Sản phẩm:", metadata.get("title", "Không rõ"))
            print("Giá:", metadata.get("selling_price", "Không rõ"))
            print("Mô tả:", metadata.get("description", "Không rõ"))
            print("Hình ảnh:", metadata.get("images", "Không rõ"))
            print("Danh mục:", metadata.get("category", "Không rõ"))
            print("----------")
    else:
        print("Không tìm thấy sản phẩm phù hợp với yêu cầu.")
```
